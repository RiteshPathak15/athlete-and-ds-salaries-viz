```python
top_10_countries=athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Filter data for women and men in the Summer Olympics
womenOlympics = athletes_df[(athletes_df['Sex'] == 'F') & (athletes_df['Season'] == 'Summer')]
menOlympics = athletes_df[(athletes_df['Sex'] == 'M') & (athletes_df['Season'] == 'Summer')]

plt.figure(figsize=(20, 15))

# Subplot 1: Women's participation over the years
plt.subplot(2, 1, 1)  # Create the first subplot in a 2x1 grid
sns.countplot(x='Year', data=womenOlympics, palette='Spectral')
plt.title("Women's Participation Over the Years")

# Subplot 2: Men's participation over the years
plt.subplot(2, 1, 2)
sns.countplot(x='Year', data=menOlympics, palette='Spectral')
plt.title("Men's Participation Over the Years")
plt.tight_layout()
```

### **Resultant Observation from above**
*   Women Participants over the year is increasing in olympics
*   Men Participants over the year is balanced in olympics

# **Bivariate Analysis**

```python
# Bivariate analysis example: Age vs. Weight
sns.scatterplot(x="Age", y="Weight", data=athletes_df)
plt.title("Bivariate Analysis: Age vs. Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()

# Bivariate analysis example: Sex vs. Medal
cross_tab = pd.crosstab(athletes_df["Sex"], athletes_df["Medal"])
cross_tab.plot(kind="bar", stacked=True)
plt.title("Bivariate Analysis: Sex vs. Medal")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

# Bivariate analysis example: Year vs. Medal
year_medal_counts = athletes_df.groupby(["Year", "Medal"]).size().unstack(fill_value=0)
year_medal_counts.plot(kind="bar", stacked=True)
plt.title("Bivariate Analysis: Year vs. Medal")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

## **Resultant Observation from above**
**1. Age vs. Weight:**
* This scatterplot visualizes the relationship between age and weight of athletes. Each point represents an athlete, with age on the x-axis and weight on the y-axis. Scatterplots like this help you understand if there's any correlation or pattern between age and weight among athletes.

**2. Sex vs. Medal:**
* This bar chart shows a cross-tabulation (crosstab) of sex (Male and Female) versus medal type (Gold, Silver, Bronze). Each bar represents the count of medals won by each sex. It helps you compare the distribution of medals between males and females.

**3. Year vs. Medal:**
* This stacked bar chart displays the count of medals (Gold, Silver, Bronze) won in each year. The bars are stacked, so you can see the total medal count for each year, broken down by medal type. It provides insights into how medal distribution has changed over the years.
