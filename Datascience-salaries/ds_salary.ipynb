# **Bivariate**

```python
result1 = ds_salary.groupby('experience_level')['salary_in_usd'].agg(['mean', 'min', 'max'])
result1['mean'] = result1['mean'].round(2)

# Resetting index to make 'experience_level' a column for Seaborn plotting
result1.reset_index(inplace=True)

# Creating subplots for three different bar plots
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plotting mean salary
sns.barplot(ax=axes[0], x='experience_level', y='mean', data=result1, color='skyblue')
axes[0].set_title('Mean Salary by Experience Level')
axes[0].set_xlabel('Experience Level')
axes[0].set_ylabel('Salary (Mean)')

# Plotting max salary
sns.barplot(ax=axes[1], x='experience_level', y='max', data=result1, color='salmon')
axes[1].set_title('Max Salary by Experience Level')
axes[1].set_xlabel('Experience Level')
axes[1].set_ylabel('Salary (Max)')

# Plotting min salary
sns.barplot(ax=axes[2], x='experience_level', y='min', data=result1, color='lightgreen')
axes[2].set_title('Min Salary by Experience Level')
axes[2].set_xlabel('Experience Level')
axes[2].set_ylabel('Salary (Min)')

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

#**1.Mean Salary by Experience Level:**
* The first subplot (axes[0]) displays the mean salary for each experience level.
* The "Senior-level" experience category tends to have the highest mean salary, followed by the "Executive-level." This suggests that, on average, individuals with more experience or in higher-level roles receive higher salaries.

#**2.Max Salary by Experience Level:**
*   The second subplot (axes[1]) illustrates the maximum salary for each
experience level.
*   Similar to the mean salary, the "Senior-level" and "Executive-level" categories show higher maximum salaries, indicating that some individuals in these roles command higher compensation.

# **3.Min Salary by Experience Level:**
*   The third subplot (axes[2]) presents the minimum salary for each experience level.
*   The "Entry-level" category generally has the lowest minimum salary, which is expected as entry-level positions typically offer lower starting salaries.


```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='experience_level', y='salary_in_usd', data=ds_salary, hue='employment_type', palette='Set2')
plt.title('Salary Distribution by Experience Level and Employment Type')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Employment Type')
plt.show()
```

```python
# Assuming 'ds_salary' is your DataFrame
df_2023 = ds_salary[ds_salary['work_year'] == 2023]
top_countries = pd.DataFrame(df_2023['company_location'].value_counts(normalize=True).head())
total_top_countries = top_countries.sum()
top_countries.loc['Others'] = 1 - total_top_countries

ax = top_countries.plot(kind='bar', legend=False, color='skyblue')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.set_ylabel('% of total jobs')
ax.set_xlabel('Company Location')
ax.set_title('Distribution of Jobs by Location in 2023', fontsize=15)
plt.show()
```

# **Observation:**
The bar plot displays the percentage distribution of jobs across the top company locations in the year 2023.
Locations are ranked by the number of jobs, and the top locations are represented in the bar plot.In which US has more job compared to other country

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'job_satisfaction' is a column in your DataFrame
top_10_job_titles = ds_salary['job_title'].value_counts().nlargest(10).index
filtered_data = ds_salary[ds_salary['job_title'].isin(top_10_job_titles)]

plt.figure(figsize=(12, 8))
sns.barplot(x='job_title', y='experience_level', data=filtered_data, color='blue', orient='h')
plt.title('Bivariate Horizontal Bar Plot of Job Title vs. Years of Experience (Top Job Titles)')
plt.xlabel('Job Title')
plt.ylabel('Years of Experience')
plt.show()
```

# **Multivariate**

```python
import seaborn as sns

# Choose specific variables for the pair plot
pairplot_variables = ['work_year', 'experience_level', 'salary_in_usd']

# Create a pair plot
sns.pairplot(ds_salary[pairplot_variables], hue='experience_level', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Work Year, Experience Level, and Salary', y=1.02)
plt.show()
```

# **Work Year vs. Salary:**
As the 'work_year' increases, there may be patterns or trends in how 'salary_in_usd' changes. The scatterplots on the diagonal axis show the univariate distributions of each variable.

# **Experience Level vs. Salary:**
The scatterplots off the diagonal axes show the bivariate relationships. The hue ('experience_level') allows you to see how different experience levels are distributed for the given combinations of 'work_year' and 'salary_in_usd'.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 1x2 grid for subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Subplot 1: Salary Comparison Across Experience Levels and Employment Types
sns.barplot(x='experience_level', y='salary_in_usd', data=ds_salary, hue='employment_type', ci=None, palette='Set2', ax=axes[0])
axes[0].set_title('Salary Comparison Across Experience Levels and Employment Types')
axes[0].set_xlabel('Experience Level')
axes[0].set_ylabel('Salary (USD)')
axes[0].legend(title='Employment Type')

# Subplot 2: Salary Distribution Over Years
sns.boxplot(x='work_year', y='salary_in_usd', data=ds_salary, palette='coolwarm', ax=axes[1])
axes[1].set_title('Salary Distribution Over Years')
axes[1].set_xlabel('Work Year')
axes[1].set_ylabel('Salary (USD)')

# Adjust layout
plt.tight_layout()
plt.show()
```

# **General Observations:**

* Subplot 1 allows for a comparison of average salaries among different experience levels and employment types.
*  The boxplot in Subplot 2 can help identify trends or variations in salary distribution over the years.
